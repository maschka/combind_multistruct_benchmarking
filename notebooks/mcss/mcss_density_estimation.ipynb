# Density Estimates of unnormalized

Hopefully with more data sparse regions will smooth out, but if not saturate at some value.

```python
flat = np.vstack([[r1 < 2.0 and r2 < 2.0 for rmsds1, rmsds2 in zip(rmsd1, rmsd2) for r1, r2 in zip(rmsds1, rmsds2)],
                  [__m for _m in mcss for __m in _m ]
                 ]).T

native = DensityEstimate(reflect=True, domain = (0, 15), points = 500, sd = 0.07*15).fit(flat[flat[:, 0]==1, 1])
decoy = DensityEstimate(reflect=True, domain = (0, 15),  points = 500, sd = 0.07*15).fit(flat[flat[:, 0]==0, 1])

plt.plot(native.x, native.fx)
plt.plot(decoy.x, decoy.fx)
plt.show()

ratio = native.ratio(decoy)
plt.plot(ratio.x, ratio.fx)
plt.show()
plt.plot(ratio.x, -np.log(ratio.fx))
plt.ylim(-10, 10)
plt.show()
```
