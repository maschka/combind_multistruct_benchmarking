```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
X = []
with open('../../../mcss.txt') as fp:
    for line in fp:
        _, _, l1, l2, m, *rest = line.strip().split(',')
        X += [list(map(int, [l1, l2, m]))]
X = np.array(X)
```

```python
plt.hist(X[:, :2], bins = 20)
plt.show()
```

```python
plt.hist(X[:, 2], bins = 20)
plt.show()
```

```python
plt.hist(X[:, 2] / X[:, :2].min(axis = 1), bins = 20)
plt.show()
```

```python
plt.hist(X[:, 2] / X[:, :2].max(axis = 1), bins = 20)
plt.show()
```

```python
plt.hist(X[(X[:, 2] / X[:, :2].min(axis = 1)) > 0.5, 2], bins = 20)
plt.show()
```

```python
plt.hist(X[(X[:, 2] / X[:, :2].max(axis = 1)) > 0.5, 2], bins = 20)
plt.show()
```

```python
print(sum((X[:, 2] / X[:, :2].min(axis = 1)) > 0.5))
print(sum((X[:, 2] / X[:, :2].max(axis = 1)) > 0.5))
```

```python
y = [sum(((X[:, 2] / X[:, :2].min(axis = 1)) > 0.5) * (X[:, 2] > i))
     for i in range(30)]
plt.plot(y)
plt.show()
```

```python
y = [sum(((X[:, 2] / X[:, :2].max(axis = 1)) > 0.5) * (X[:, 2] > i))
     for i in range(30)]
plt.plot(y)
plt.show()
```

```python
current  = ((X[:, 2] / X[:, :2].min(axis = 1)) > 0.5) * (X[:, 2] > 10)

proposed = ((X[:, 2] / X[:, :2].max(axis = 1)) > 0.5) * (X[:, 2] > 10)

print(sum(current), sum(proposed))
print(sum(current*proposed))
```
